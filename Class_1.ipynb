**Classification**

* Evaluation is based e.g. on misclassification (or error) rate

**Regression**

### Learning Algorithms
* **Decision Stump**  
    A simple decision tree with only one split (one-level tree). It selects a single feature and threshold to make a decision. Often used as a weak learner in ensemble methods like AdaBoost.

* **KNN (K-Nearest Neighbors)**  
    A non-parametric method that classifies a sample based on the majority label among its $k$ closest neighbors in the feature space. For regression, it averages the values of the $k$ nearest neighbors.

* **Decision Tree**  
    A model that recursively splits the data based on feature values to create a tree structure. Each internal node represents a decision on a feature, and each leaf node represents a class label (classification) or value (regression).

* **SVM (Support Vector Machine)**  
    A powerful classifier that finds the optimal hyperplane separating classes with the maximum margin. Can be extended to non-linear boundaries using kernel functions.

* **Linear Regression**  
    A regression algorithm that models the relationship between input features and a continuous output by fitting a linear equation to the observed data. For multiple features, it is called multiple linear regression.

* **Naive Bayes**  
    A probabilistic classifier based on Bayes' theorem, assuming independence between features. It is simple, fast, and effective for high-dimensional data, commonly used in text classification and spam detection.

* **RNN (Recurrent Neural Network)**  
    A type of neural network designed for sequential data, where connections between nodes form directed cycles. RNNs are well-suited for tasks involving time series, natural language, and other data where context and order are important.  

* **Unsupervised Learning**  
    In unsupervised learning, the algorithm is provided with input data $X$ without any corresponding output labels $y$. The goal is to discover hidden patterns, structures, or relationships within the data. Common tasks include:
    - **Clustering**: Grouping similar data points together based on feature similarity (e.g., K-Means, Hierarchical Clustering).
    - **Dimensionality Reduction**: Reducing the number of features while preserving important information (e.g., Principal Component Analysis (PCA), t-SNE).
    - **Anomaly Detection**: Identifying data points that deviate significantly from the majority.
    Unsupervised learning is useful when labeled data is scarce or expensive to obtain, and it helps in exploratory data analysis, feature engineering, and data compression.

* **Supervised Learning**  
    In supervised learning, the algorithm is trained on a labeled dataset, meaning each input $x_i$ is paired with a corresponding output $y_i$. The objective is to learn a mapping from inputs to outputs that generalizes well to unseen data. Supervised learning is divided into:
    - **Classification**: Predicting discrete class labels (e.g., spam detection, image recognition).
    - **Regression**: Predicting continuous values (e.g., house price prediction).
    The learning process involves minimizing a loss function that measures the difference between predicted and true outputs. Supervised learning is widely used in applications where historical labeled data is available.

* **Reinforcement Learning**  
    Reinforcement learning (RL) is a paradigm where an agent interacts with an environment by taking actions and receiving feedback in the form of rewards or penalties. The agent's goal is to learn a policy (a strategy for choosing actions) that maximizes the cumulative reward over time. Key concepts include:
    - **Agent**: The learner or decision-maker.
    - **Environment**: The system with which the agent interacts.
    - **State**: The current situation of the agent in the environment.
    - **Action**: Choices available to the agent.
    - **Reward**: Feedback signal indicating the immediate benefit of an action.
    RL is used in areas such as robotics, game playing (e.g., AlphaGo), recommendation systems, and autonomous vehicles, where learning from direct experience and delayed feedback is essential. Unlike supervised learning, RL does not require labeled input/output pairs but instead relies on trial-and-error exploration.


Principal Component Analysis (PCA)
----------------------------------

Principal Component Analysis (PCA) is a dimensionality reduction technique used to simplify complex datasets by transforming them into a new set of variables called **principal components**. These components are linear combinations of the original features and are ordered so that the first few retain most of the variation present in the original dataset.

### Why use PCA?
- **Reduce dimensionality**: Lower the number of features while preserving as much information as possible.
- **Visualization**: Project high-dimensional data into 2D or 3D for visualization.
- **Noise reduction**: Remove less informative (noisy) components.

### How PCA Works

1. **Standardize the Data**  
    Center the data by subtracting the mean and (optionally) scale to unit variance. This ensures all features contribute equally.

2. **Compute the Covariance Matrix**  
    Calculate the covariance matrix to understand how variables relate to each other.

3. **Calculate Eigenvectors and Eigenvalues**  
    Find the eigenvectors (principal components) and their corresponding eigenvalues of the covariance matrix.  
    - **Eigenvectors** define the directions of the new feature space.
    - **Eigenvalues** indicate the amount of variance carried in each principal component.

4. **Sort and Select Principal Components**  
    Sort the eigenvectors by their eigenvalues in descending order. Select the top $k$ eigenvectors to form a new feature space (where $k$ is the desired number of dimensions).

5. **Transform the Data**  
    Project the original data onto the selected principal components to obtain the reduced representation.

### Summary

PCA finds the axes (principal components) along which the data varies the most and projects the data onto these axes, reducing dimensionality while retaining as much variance as possible. It is widely used for exploratory data analysis, visualization, and as a preprocessing step for other machine learning algorithms.


### Simple Classifiers 
A **Decision Stump** is a simple classifier that consists of a single-level decision tree. It selects one feature and a threshold to split the data into two groups, making a decision based on whether the feature value is above or below the threshold.

**Algorithm to Find the Threshold:**
1. For each feature, sort the data points by their values.
2. For each possible threshold (typically midpoints between consecutive sorted values), evaluate the classification error if the data is split at that threshold.
3. For each threshold, assign one class to values below the threshold and the other class to values above (and vice versa), and compute the error for both assignments.
4. Select the feature and threshold that yield the lowest classification error.

This process results in a simple rule: "If feature $x_j < t$, predict class A; else, predict class B." Decision stumps are often used as weak learners in ensemble methods like AdaBoost.



```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Load a simple dataset (Iris)
X, y = load_iris(return_X_y=True)
# For binary classification, select only two classes
X, y = X[y < 2], y[y < 2]

# Split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

# Train a decision stump with max_depth=1
stump = DecisionTreeClassifier(max_depth=1, random_state=42)
stump.fit(X_train, y_train)

# Predict and evaluate
y_pred = stump.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
```

## Decision Tree

A **Decision Tree** is a supervised learning algorithm used for both classification and regression tasks. It works by recursively splitting the data into subsets based on feature values, creating a tree-like structure where each internal node represents a decision on a feature, each branch represents the outcome of the decision, and each leaf node represents a class label (for classification) or a value (for regression).

---

### Gini Criterion

The **Gini impurity** is a measure of how often a randomly chosen element from the set would be incorrectly labeled if it was randomly labeled according to the distribution of labels in the subset. It is used to evaluate splits in the data:


$G_i = 1 - \sum_{k=1}^n p_{i,k}^2$

where \( p_{i,k} \) is the probability of class \( k \) at node \( i \).

- **Lower Gini impurity** means a purer node (more of one class).
- Used in algorithms like CART (Classification and Regression Trees).
- The split is chosen to minimize the Gini impurity in the resulting nodes.

---

### Orthogonal Hyperplanes

Decision trees split the feature space using **hyperplanes** that are always orthogonal (perpendicular) to the feature axes. This means each split only considers one feature at a time, resulting in axis-aligned partitions.

---

### White Box vs Black Box Models

- **White Box Models**: Models like decision trees are interpretable and transparent; you can easily understand how decisions are made.
- **Black Box Models**: Models like neural networks or ensemble methods (e.g., random forests) are complex and less interpretable.

---

### Learning Methods

- **CART (Classification and Regression Trees)**: Uses Gini impurity for classification and mean squared error for regression.
- **ID3 (Iterative Dichotomiser 3)**: Uses information gain (based on entropy) to select splits.
- **Quinlan C4.5**: An extension of ID3 that handles both continuous and discrete attributes, uses gain ratio, and can handle missing values.

---

### Gini Calculation

The Gini impurity transforms class counts into a probability distribution and measures the "impurity" of a node. A node is pure (Gini = 0) if all samples belong to one class.

---

### Shannon Entropy

**Entropy** measures the amount of uncertainty or impurity in the data:


$I = -\sum_{k=1}^n p_k \log_2 p_k$

- \( I \) is measured in **bits** (the information unit).
- Higher entropy means more disorder (mixed classes).
- Used in ID3 and C4.5 to select the best split (maximizing information gain).

---

### Cross Entropy

**Cross entropy** measures the difference between two probability distributions (the true distribution and the predicted distribution). It is widely used as a loss function in classification tasks.

---

### Bits vs Nats

- **Bits**: Use logarithm base 2 (\( \log_2 \)), common in information theory.
- **Nats**: Use natural logarithm (\( \log_e \)), common in mathematics and physics.

---

**Summary:**  
Decision trees are interpretable models that split data using simple rules based on features. They use criteria like Gini impurity or entropy to choose the best splits, and their structure makes them easy to visualize and understand.

```python
# Use only the first two features (2D)
X_train_2d = X_train[:, :2]
X_test_2d = X_test[:, :2]

# Train a decision tree classifier (no depth limit) on 2D data
tree_2d = DecisionTreeClassifier(random_state=42)
tree_2d.fit(X_train_2d, y_train)

# Predict and evaluate
y_pred_tree_2d = tree_2d.predict(X_test_2d)
print("Decision Tree Accuracy (2D):", accuracy_score(y_test, y_pred_tree_2d))
print("Number of features (dimensions):", X_train_2d.shape[1])

# Visualize the decision boundaries (optional, requires matplotlib)
import matplotlib.pyplot as plt 
def plot_decision_boundary(clf, X, y):
    import numpy as np
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01),
                         np.arange(y_min, y_max, 0.01))
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    
    plt.contourf(xx, yy, Z, alpha=0.8)
    plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', marker='o')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.title('Decision Boundary')
    plt.show()
# Plot the decision boundary for the 2D decision tree
plot_decision_boundary(tree_2d, X_train_2d, y_train)

# Print tree depth and number of leaves
print("Tree depth:", tree_2d.get_depth())
print("Number of leaves:", tree_2d.get_n_leaves())

# Optionally, print the feature importances
print("Feature importances:", tree_2d.feature_importances_)


```
